```python
import pandas as pd
df = pd.DataFrame({'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C'],
                   'position': ['A-1', 'A-1', 'G-1', 'G-1', 'B-1', 'B-2', 'H-1', 'H-2', 'K-1', 'K-1', 'K-2'],
                   'points': [25, 17, 14, 9, 12, 9, 6, 4, 8, 12, 6]})

df
```

```python
def return_multi_group(df, level = 'team', sub_level =  'position', col = 'points'):

    g1 = df.groupby([level, sub_level])[col].sum().reset_index()

    mask = g1[level].ne(g1[level].shift(-1))
    g11 = pd.DataFrame('',index=mask.index[mask] + .5, columns=g1.columns)
    g1 = pd.concat([g1, g11]).sort_index().reset_index(drop=True).iloc[:-1]

    g2 = df.groupby([level])[col].sum().reset_index()
    g2[level] = 'SUM_' + g2[level].astype(str)
    g2.loc[:,sub_level] = ''
    g2 = g2[[level, sub_level, col ]]

    j = 0
    for i in range(len(g1)):
        if g1.loc[i,level] == '':
            j = j + 1
            g1.loc[i, : ] = list(g2.T.pop(j - 1))

    g1.loc[len(g1),:] = list(g2.T.pop(j))

    return g1




```

```python
return_multi_group(df)
```
